```python
# set the id 
data.set_index("id", inplace=True)
data.head()
```

```python
# which columns to go for? 
data.columns
```

```python
# lets pick mean value columns 
# doing y = mx+c 
x = data.iloc[:, 1:11]

x
```

```python
y = data.loc[:, ["diagnosis"]]
y
```

Divide the data into Training and Testing 

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2)
```

```python
ytrain
```

```python
# mu and sigma [ mean and standard deviation ]
xtrain.describe().T
```

```python
statistics = xtrain.describe().T
xmean = statistics["mean"]
xstandard_deviation = statistics["std"]
xmean, xstandard_deviation
```

```python
normalised_xtrain = (xtrain - xmean) / xstandard_deviation
normalised_xtest = (xtest - xmean) / xstandard_deviation
```

```python
normalised_xtest.head()
```

```python
ytrain.head()
```

Our Y is M and B ML only understands 0 and 1 
M => Cancer => 1 
B => Cancer => 0 

```python
convert_diagnosis = lambda d: 1 if d=="M" else 0
ytrain_encoded = ytrain["diagnosis"].map(convert_diagnosis)
ytest_encoded = ytest["diagnosis"].map(convert_diagnosis)
```

```python
print(ytrain_encoded.tail(10))
```

```python
# x values are Normalised 
# y values are INT and are nolonger strings 
# DATA is ready for Machine Learning 
```

```python
# Define the Algorithms
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

# train the model [training]
model.fit(normalised_xtrain, ytrain_encoded)

# testing/scoring 
predictions = model.predict(normalised_xtest)

# evaluations 
from sklearn.metrics import accuracy_score
accuracy_score(predictions, ytest_encoded)
```

```python
from sklearn.naive_bayes import GaussianNB
model2 = GaussianNB()
# train the model 
model2.fit(normalised_xtrain, ytrain_encoded)
# training and scoring 
predictions = model2.predict(normalised_xtest)
# evaluations 
from sklearn.metrics import accuracy_score
accuracy_score(predictions, ytest_encoded)
```

```python
# Use different Algorithms 
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC

# Hyperparameters two different hyperameters for Random Forest 
# Same Algorithm can also 2 different models 

ntree1, ntree2 = 50, 100
for data in [RandomForestClassifier(n_estimators = ntree1), RandomForestClassifier(n_estimators=ntree2), GaussianNB(), DecisionTreeClassifier(), KNeighborsClassifier(), LinearSVC()]:
    model = data
    model.fit(normalised_xtrain, ytrain_encoded)
    predictions = model.predict(normalised_xtest)
    print(accuracy_score(predictions, ytest_encoded))
```
